# Ejemplo 1


Consideramos el 3-símplice 
(0,1,2,3)

```python
sc=SimplicialComplex([(0,1,2,3)])
```

# Conjunto de todas sus caras

```python
sc.face_set
```

# Dimensión

```python
sc.dimension
```

# Conjunto de vértices

```python
sc.n_faces(0)
```

# Conjunto de aristas

```python
sc.n_faces(1)
```

# Conjunto de 2-símplices

```python
sc.n_faces(2)
```

# Conjunto de 3-símplices

```python
sc.n_faces(3)
```

# Característica de Euler

```python
sc.Euler_characteristic
```

# Estrella de la arista (0,1)

```python
sc.st((0,1))
```

# Link de la arista (0,1)

```python
sc.lk((0,1))
```

# Número de componentes conexas

```python
sc.connected_components()
```
